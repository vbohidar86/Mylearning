```python
league=league.astype({'GoalsFor':'int64','GoalsAgainst':'int64'})
```

```python
def Goal_diff_count (league):
    l= []
    for i in league.index:
        goal_diff= int(league.iloc[i,8:9])-int(league.iloc[i,9:10])
        l.append(abs(goal_diff))
    goal_diff_df=pd.DataFrame({'Team':list(league['Team']),'Goal_diff_count':l})
    return goal_diff_df
Gd_diff=Goal_diff_count(league)
Gd_diff
```

```python
Gd_diff.min()
```

```python
Gd_diff.max()
```

```python
league=league.astype({'GamesWon':'int64','GamesPlayed':'int64'})
league.dtypes
```

```python
league['Percentage of winning']=(league['GamesWon']/league['GamesPlayed']*100)
league['Percentage of winning'].fillna(0, inplace=True)
```
